First, open a terminal window and install the package `overpy` with the command `pip install overpy`.

Import the package `overpy`

```python
import overpy
```

Initiate the overpass API

```python
api = overpy.Overpass()
```

```python
personas = ["innenstadt", "aussenstadt", "familie", "kinder_vorschule", "kinder_schule",
            "student", "single", "senior"]
```

```python
#freizeit
trinken = ["amenity=bar", "amenity=biergarten", "amenity=pub"]
essen = ["amenity=fast_food", "amenity=restaurant", "amenity=food_court"]
snack = ["amenity=cafe", "amenity=ice_cream"]
spazieren = ["leisure=park", "leisure=dog_park"]
ausgehen = ["leisure=bowlong_alley", "amenity=cinema", "amenity=escape_game",
            "leisure=water_park", "leisure=garden", "amenity=bbq"]
nightlife = ["leisure=dance", "amenity=nighclub"]
kultur = ["amenity=public_bookcase", "amenity=arts_centre", "amenity=theatre", "amenity=library"]

#sport
indoor = ["leisure=fitness_centre", "leisure=sports_centre", "leisure=swimming_pool", "leisure=water_park",
          "leisure=ice_rink"]
outdoor = ["leisure=park", "leisure=swimming_area", "leisure=pitch", "leisure=fitness_station"]

#einkaufen
einkauf = ["amenity=marketplace", "shop=beverages", "shop=supermarket"]

#bildung
vorschule = ["amenity=kindergarten", "leisure=playground"]
schule = ["amenity=college", "amenity=school", "amenity=music_school"]
uni = ["amenity=university", "amenity=research_institute"]

#mobilit√§t
fahrrad = ["amenity=bicycle_parking", "amenity=bicycle_repear_station", "amenity=bicycle_rental"]
oev = ["amenity=bus_station"]
auto = ["amenity=parking"]

#healthcare
ersteHilfe = ["amenity=clinic", "amenity=doctors", "amenity=hospital", "amenity=pharmacy", "amenity="]

#misc
finance = ["amenity=bank", "amenity=atm"]
security = ["amenity=police", "amenity=fire_station"]
post = ["amenity=post_office", "amenity=post_box"]
#ruhe = ["highway=living_street"]
```

```python
#define personas
single = [trinken, ausgehen, nightlife, einkauf, oev, indoor]
familie = [essen, spazieren, kultur, einkauf, outdoor, indoor]
student = [uni, einkauf, fahrrad, oev, ausgehen, nightlife, trinken, ausgehen]
kinder_vorschule = [vorschule, snack, fahrrad, auto, security]
kinder_schule = [schule, fahrrad, outdoor]
senior = [security, ersteHilfe, post, finance, einkauf, spazieren, snack]

#define your persona
user = "student"
anfrage = []

#define query
if user==personas[5]:
    print("jes")
    for i in student:
        for j in i:
            anfrage.append(j)
        

print(anfrage)
print(len(anfrage))
```

Construct query text

```python
#query
query_text = []
for i in anfrage:
    query = "node[" + i + "](" + f'{bi_coord["south"]:.9f}' + "," + f'{bi_coord["west"]:.9f}' + "," + f'{bi_coord["north"]:.9f}' +  "," + f'{bi_coord["east"]:.9f}' + ");out;"
    query_text.append(query)

print(query_text)

results = []
for i in query_text:
    results.append(api.query(i))
    

```

Query the overpass API for the result

Get number of retrieved nodes

```python
for i in results:
    for j in i.nodes:
        print(j)
```

Get latitude and longitude of first node

```python
import csv
import datetime

now = datetime.datetime.now()

#write results to csv
filename = str(now)[:10] + "_" + str(user) + ".csv"
print(filename)

with open(filename, mode='w', newline='') as file:
    user_writer = csv.writer(file, delimiter=';')
    for i in results:
        for j in i.nodes:
            user_writer.writerow([str(j.id), str(j.lat), str(j.lon)])
```
